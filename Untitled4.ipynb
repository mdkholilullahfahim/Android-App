```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
```

```python
agglomerative = AgglomerativeClustering(n_clusters=4)
agglomerative_labels = agglomerative.fit_predict(X)
```

```python
kmeans_score = silhouette_score(X, kmeans_labels)
agglomerative_score = silhouette_score(X, agglomerative_labels)
```

```python
print(f"K-Means Silhouette Score: {kmeans_score}")
print(f"Agglomerative Silhouette Score: {agglomerative_score}")
```

```python
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', edgecolors='k')
plt.title('K-Means Clustering')
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=agglomerative_labels, cmap='viridis', edgecolors='k')
plt.title('Agglomerative Clustering')
```

```python
plt.show()
```
